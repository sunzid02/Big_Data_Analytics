````python
#Import packages and Dataset
import numpy as np
import pandas as pd

fifa20Data = pd.read_csv("data/fifa_20.csv")#Change path according to your files

fifa20 = fifa20Data.copy(deep=False)

fifa20.head(5)
````

Take the following SQL-Queries and use Pandas to achieve the same result.

### First Query
```sql
SELECT short_name, overall, value_eur, wage_eur 
FROM fifa_20 
WHERE overall > 80 and club = 'Borussia Dortmund'
``` 

````python
#Recreate the first query with pandas
firstQuery = fifa20[['short_name', 'overall', 'value_eur', 'wage_eur']][(fifa20['overall'] > 80) & (fifa20['club'] == 'Borussia Dortmund')]

firstQuery
````

### Second Query
```sql
SELECT short_name, overall 
FROM fifa_20 
WHERE club in ('Borussia Dortmund','FC Bayern München') and nationality = 'France'
``` 

````python
#Recreate the second query with pandas
secondQuery = fifa20[['short_name', 'overall']][ fifa20['club'].isin(['Borussia Dortmund', 'FC Bayern München']) & (fifa20['nationality'] == 'France') ]

secondQuery
````

### Third Query
```sql
SELECT nationality, count(*) 
FROM fifa_20 
GROUP BY nationality 
ORDER BY count
``` 

````python
#Recreate the third query with pandas
thirdQuery =  fifa20[['nationality']].groupby('nationality').size().sort_values(ascending=False)

thirdQuery
````

### Fourth Query
```sql
SELECT club, count(*) 
FROM fifa20 
WHERE potential > 85 
GROUP BY club having count(*) > 10
ORDER BY count(*) desc
``` 

````python
#Recreate the fourth query with pandas
fourthQuery1 = fifa20[['club']][(fifa20['potential'] > 85 )].groupby('club').size()

fourthQuery2 = pd.DataFrame({'club':fourthQuery1.index, 'count':fourthQuery1.values})

fourthQuery3 = fourthQuery2[fourthQuery2["count"] > 10]
 
final_query = fourthQuery3.sort_values(by=['count'], ascending=False)

final_query
````

### Fifth Query
For the last query we create two additional dataframes that you have to join

````python
#Do NOT change this code 
player_values = fifa20[['short_name','club','age','height_cm','weight_kg']]
fifa_club = fifa20[['short_name','club','overall','potential']]
````

```sql
SELECT short_name, age, overall 
FROM player_values JOIN fifa_club on fifa_club.short_name = player_values.short_name
WHERE fifa_club.club == 'Borussia Dortmund'
``` 

````python
#Recreate the fifth query with pandas
data = pd.merge(player_values, fifa_club, on = 'short_name', how='inner').drop('club_y', axis=1)
data.rename(columns={'club_x':'club'}, inplace=True)


data[['short_name', 'age', 'overall']][(data['club'] == 'Borussia Dortmund')]
````
